```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'cleaned-landslide-data.csv'  # Replace with your file path
landslide_data = pd.read_csv(file_path)

# Grouping the data by country and counting the number of landslides in each country
country_landslide_counts = landslide_data['country_name'].value_counts()

# Selecting the top 20 countries with the most landslides
top_20_countries = country_landslide_counts.head(20)

# Creating a bar plot for the top 20 countries
plt.figure(figsize=(12, 8))
sns.barplot(x=top_20_countries.values, y=top_20_countries.index, palette="viridis")
plt.title('Top 20 Countries with Most Landslides Recorded')
plt.xlabel('Number of Landslides')
plt.ylabel('Country')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
population_90th_percentile = landslide_data['admin_division_population'].quantile(0.90)
densely_populated_data = landslide_data[landslide_data['admin_division_population'] > population_90th_percentile]
```

```python
heatmap_data = pd.crosstab(densely_populated_data['landslide_category'], densely_populated_data['landslide_trigger'])
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)
plt.title('Relationship Between Landslide Types and Triggers in Densely Populated Areas')
plt.xlabel('Landslide Trigger')
plt.ylabel('Landslide Type')
plt.show()
```
